```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('billboard_lyrics_1964-2017.csv')
```

```python
#make total words
df['Total Words'] = ""
for i in range(len(df)):
    total_words = 0
    if not(isinstance(df.iloc[i].Lyrics, float)): 
        lyrics = (df.iloc[i].Lyrics).split(' ')
        total_words = len(lyrics)
    df.loc[df.index[i], 'Total Words'] = total_words
    
for i in range(len(df)):
    print(df.loc[df.index[i], 'Total Words'])
```

```python
#make unique words
df['Unique Words'] = ""
for i in range(len(df)):
    unique_words = set()
    if not(isinstance(df.iloc[i].Lyrics, float)): 
        lyrics = (df.iloc[i].Lyrics).split(' ')
        for word in lyrics:
            unique_words.add(word)
    df.loc[df.index[i], 'Unique Words'] = len(unique_words)
    
for i in range(len(df)):
    print(int(df.loc[df.index[i], 'Unique Words']))
```

```python
songs_per_year = 100
year_start = 1965
year_end = 2017

overlap_top = np.zeros((0,3))
for year in range(year_start, year_start + 2):
    overlap = np.zeros((songs_per_year, songs_per_year))
    for i in range(songs_per_year):
        adj_i = (year - year_start) + i
        for j in range(songs_per_year):
            adj_j = (year - year_start) + j
            if (i != j):
                sharedWords = set()
                if (isinstance(df.iloc[adj_i].Lyrics, float) or isinstance(df.iloc[adj_j].Lyrics, float)):
                    overlap[i,j] = 0
                    continue
                if (df.iloc[adj_i].Lyrics == 'NA' or df.iloc[adj_j].Lyrics == 'NA'):
                    overlap[i,j] = 0
                    continue
                lyrics1 = (df.iloc[adj_i].Lyrics).split(' ')
                lyrics2 = (df.iloc[adj_j].Lyrics).split(' ')
                for word in lyrics1:
                    if (word in lyrics2):
                        sharedWords.add(word)
                overlap[i,j] = len(sharedWords)

    num_connections = 2
    connections = set()

    for i in range(songs_per_year):
        rank = overlap[i].argsort()[-num_connections:][::-1]
        for j in range(len(rank)):
            if ((i, rank[j]) not in connections):
                overlap_top = np.vstack((overlap_top, [year, i, rank[j]]))
            connections.add((i, rank[j]))
            connections.add((rank[j], i))
        
np.savetxt("overlap_test.csv", np.vstack((np.arange(songs_per_year), overlap)), delimiter=",")
overlap_top = np.vstack((['Year','Start', 'End'], overlap_top))
np.savetxt("overlap_top_test.csv", overlap_top, delimiter=",", fmt="%s")
```
